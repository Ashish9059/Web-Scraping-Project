# Wikipedia Article Scraping

###### In this project we will take a person’s name as input from the user and then scrape his/her Wikipedia page. All the scraped information will be written in a text file and then saved in the current working directory of the user. 

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://en.wikipedia.org/wiki/Martin_Luther_King_Jr.')
soup = BeautifulSoup(res.text,'html.parser')
```

```python
heading = soup.find('h1').text
```

```python
heading
```

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'
    
corpus = corpus.strip()
```

```python
print(corpus)
```

```python
fd = open(heading + '.txt', 'w', encoding = 'utf-8')
fd.write(corpus)
fd.close()
```

# Wikipedia Scraping by Title

```python
import requests
from bs4 import BeautifulSoup
```

```python
title = str(input('Enter the topic: ')).replace(' ','+')
link = 'https://www.google.com/search?q=' + title + '+wikipedia'

res = requests.get(link)
soup = BeautifulSoup(res.text,'html.parser')


for sp in soup.find_all('div'):
    try:
        link = sp.find('a').get('href')
        if ('en.wikipedia.org' in link):
            break
    except:
        pass
    
link = link[7:].split('&')[0]
```

```python
res = requests.get(link)
soup = BeautifulSoup(res.text,'html.parser')

corpus = ''
for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'
    
corpus = corpus.strip()

for i in range(500):
    corpus = corpus.replace('['+str(i)+']','')
```

```python
print(corpus)
```
